```
import matplotlib.pyplot as plt
age=[21,25,30,35,40,45,50]
salary=[20000,25000,30000,35000,40000,45000,50000]
plt.plot(age, salary, label="Salary on basis of Experience", color='green')

plt.legend()
plt.xlabel("age")
plt.ylabel("salary")
plt.show()
```
